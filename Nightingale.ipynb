#### Importing all the required libraries

```
import plotly.graph_objects as go
import pandas as pd
import plotly.tools
```

#### Reading the Nightingale dataset

```
df_nightingale = pd.read_excel("/content/drive/My Drive/Data Visualisation/nightingale-data.xlsx", header=1);
```

```
df_nightingale
```

#### Plotting BarPolar Chart based on Annual Mortality Rate from April 1855 to March 1856

```
fig= go.Figure()
fig.add_trace(go.Barpolar(
    r= df_nightingale["All other causes.1"][12:],
    theta = df_nightingale["Month"][12:],
    name="All other causes",
    marker=dict(color='rgb(0, 255, 255)', opacity=0.5)
))
fig.add_trace(go.Barpolar(
    r= df_nightingale["Wounds & injuries.1"][12:],
    theta = df_nightingale["Month"][12:],
    name='Wounds & injuries',
    marker=dict(color='rgb(241, 148, 138)', opacity=0.5)
))
fig.add_trace(go.Barpolar(
    r= df_nightingale["Zymotic diseases.1"][12:],
    theta = df_nightingale["Month"][12:],
    name= "Zymotic diseases",
    marker_color='rgb(144, 148, 151)' 
))
fig.update_layout(
    title='DIAGRAM of the CAUSES of MORTALITY in the ARMY in the EAST',
    font_size=14,
    legend_font_size=14,
    width=800,
    height=800,
    polar_angularaxis_direction= 'clockwise',
    polar=dict( radialaxis= dict(visible = False)),
    template ="plotly_white"
    )
```
